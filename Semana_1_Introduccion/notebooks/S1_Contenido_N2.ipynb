## Ejercicio 1: Calculadora de Propinas

Vamos a crear un programa que calcule automÃ¡ticamente la propina en un restaurante.

```python
# Calculadora de propinas para restaurante

# Datos de la cuenta
total_cuenta = float(input("Ingresa el total de la cuenta: $"))
porcentaje_propina = float(input("Â¿QuÃ© porcentaje de propina quieres dar? (ej: 15): "))
numero_personas = int(input("Â¿Entre cuÃ¡ntas personas van a dividir la cuenta?: "))

# CÃ¡lculos
propina = total_cuenta * (porcentaje_propina / 100)
total_con_propina = total_cuenta + propina
por_persona = total_con_propina / numero_personas

# Mostrar resultados con F-strings
print(f"""
=== CALCULADORA DE PROPINAS ===
Cuenta original:    ${total_cuenta:.2f}
Propina ({porcentaje_propina}%):     ${propina:.2f}
Total con propina:  ${total_con_propina:.2f}
Por persona:        ${por_persona:.2f}
===============================
""")
```
