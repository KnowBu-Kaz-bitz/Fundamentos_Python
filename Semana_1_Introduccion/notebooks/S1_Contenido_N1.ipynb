```python
# Las variables permiten guardar información y usarla después
# Cada variable puede ser de un tipo distinto
# Supongamos que queremos crear un perfil de un usuario

# str (string): Para guardar el nombre del usuario
nombre_usuario = "Peter Parker"

# int (integer): Para números enteros, como la edad o el ID
edad = 28
id_usuario = 101

# float (flotante): Para números decimales. Como una calificación o el balance de cuenta
calificacion_promedio = 9.5
balance_banco = 1202.2

# bool (booleano): Para valores verdaero (True) o Falso (False)
cliente_premium = True
auto = False

# Ahora podemos combinar la función print() y las variables para mostrar la información
print("--- Información del usuario ---\n")
print("Nombre: ", nombre_usuario)
print("Edad: ", edad)
print("Calificación: ", calificacion_promedio)
print("¿Es cliente premium?:", cliente_premium)
print("¿Cuenta con automóvil?:", auto)
print("\n-------------------------------")
```

## 3. Operadores matemáticos y concatenación
#### Objetivos del notebook.
- Objetivo 1: Conocer los tipos de operadores en Python
- Objetivo 2: Realizar operaciones con operadores y concatenaciones
- Dataset: No se requieren ningún archivo adicional

```python
# Python también permite el uso de operadores matemáticos
# Puede funcionar como una calculadora
# Vamos a simular el cálculo total de una compra.

# Definir la fecha de compra
fecha_compra = "04/11/2025"

# Precios de productos
precio_cafe = 75
precio_pan = 20

# Cantidad de productos
cantidad_cafe = 2
cantidad_pan = 4

# Calcular subtotal
subtotal_cafe = precio_cafe * cantidad_cafe
subtotal_pan = precio_pan * cantidad_pan

# Calculamos el total de la compra
total_compra = subtotal_cafe + subtotal_pan

# Realizamos un resumen de la compra
print("--- Resumen de compra ---\n")
print("Fecha de compra:", fecha_compra)
print("Subtotal café:", subtotal_cafe)
print("Subtotal pan:", subtotal_pan)
print("Total:", total_compra)

# El operador de suma puede usarse para concatenar cadena
# Vamos a mostrar el ejemplo con los datos anteriores
# Vamos a definir un nombre de usuario
nombre_usuario = "Carlos"
mensaje_bienvenida = "Hola, "

# Concatenamos el mensaje completo
mensaje_completo = mensaje_bienvenida + nombre_usuario + ". Gracias por tu compra."
print("\nMensaje para el cliente")
print(mensaje_completo)
```

## 5. Formateo de cadenas (F-strings)
#### Objetivos del notebook.
- Objetivo 1: Generar outputs de cadenas personalizados
- Objetivo 2: Combinar tipos de datos, operadores matemáticos y concatenaciones para formatear outputs
- Dataset: No se requieren ningún archivo adicional

```python
# --- Simulador de Presupuesto Mensual ---

# ** STRINGS (str) **: Almacenan texto para personalizar la experiencia.
nombre_usuario = "Alex"
mes = "Octubre"

# ** INTEGERS (int) **: Se usan para contar cosas, como los días.
dias_del_mes = 31

# ** FLOATS (float) **: Perfectos para dinero, ya que manejan decimales.
ingreso_mensual = 18500.50
gasto_renta = 7500.00
gasto_comida = 4250.75
gasto_transporte = 980.25
ahorro_adicional = 1200.00

# ** BOOLEAN (bool) **: Para registrar un estado de sí o no.
meta_ahorro_alcanzada = False # Aún no lo sabemos, lo calcularemos.

# --- CÁLCULOS Y LÓGICA ---

# Sumamos todos los gastos para obtener un total.
total_gastos = gasto_renta + gasto_comida + gasto_transporte

# Calculamos el dinero que queda después de gastos.
flujo_libre = ingreso_mensual - total_gastos

# Calculamos el ahorro final del mes.
ahorro_final = flujo_libre - ahorro_adicional

# Calculamos el promedio de gasto diario.
gasto_diario_promedio = total_gastos / dias_del_mes

# Verificamos si el ahorro final es mayor a $5,000 (nuestra meta).
if ahorro_final > 5000:
    meta_ahorro_alcanzada = True

# --- REPORTE FINAL ---
# Usamos print() con f-strings para crear un reporte financiero detallado y fácil de leer.
# La sintaxis {:.2f} formatea los números para que siempre muestren 2 decimales.
print("=====================================================")
print(f"      Reporte Financiero para {nombre_usuario} - Mes: {mes}")
print("=====================================================")
print(f"Ingreso Mensual Total:      ${ingreso_mensual:.2f} MXN")
print("-----------------------------------------------------")
print("Desglose de Gastos:")
print(f"  - Renta:                  ${gasto_renta:.2f} MXN")
print(f"  - Comida:                 ${gasto_comida:.2f} MXN")
print(f"  - Transporte:             ${gasto_transporte:.2f} MXN")
print(f"Total de Gastos Fijos:      ${total_gastos:.2f} MXN")
print("-----------------------------------------------------")
print(f"Flujo de Dinero Libre:      ${flujo_libre:.2f} MXN")
print(f"Ahorro Adicional Planeado:  ${ahorro_adicional:.2f} MXN")
print(f"Ahorro Final Real del Mes:  ${ahorro_final:.2f} MXN")
print("=====================================================")
print(f"Análisis Adicional:")
print(f"  - Gasto Diario Promedio:  ${gasto_diario_promedio:.2f} MXN")
print(f"  - ¿Se cumplió la meta de ahorro de $5,000?: {meta_ahorro_alcanzada}")
print("=====================================================")
```
