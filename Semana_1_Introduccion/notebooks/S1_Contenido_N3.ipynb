# Fundamentos de Python: Proyecto Mini

## Mini Proyecto - Sistema de Gestión Personal

### Objetivo:
Crear un sistema simple que combine todos los conceptos aprendidos:
- Variables y tipos de datos
- Operadores matemáticos
- Conversiones de tipos
- F-strings y formateo

### Descripción:
Desarrollaremos un sistema que ayude a gestionar información personal y calcular algunos datos útiles.

---

```python
# MINI PROYECTO: Sistema de Gestión Personal

print("=== SISTEMA DE GESTIÓN PERSONAL ===")
print("Bienvenido/a a tu asistente personal\n")

# 1. Recopilar información personal
nombre = input("¿Cuál es tu nombre completo?: ")
año_nacimiento = int(input("¿En qué año naciste?: "))
peso = float(input("¿Cuál es tu peso en kg?: "))
altura = float(input("¿Cuál es tu altura en metros? (ej: 1.75): "))

# 2. Recopilar información financiera básica
ingreso_mensual = float(input("¿Cuál es tu ingreso mensual?: "))
gasto_mensual = float(input("¿Cuánto gastas aproximadamente al mes?: "))

# 3. Cálculos automáticos
año_actual = 2025
edad = año_actual - año_nacimiento
imc = peso / (altura ** 2)
ahorro_mensual = ingreso_mensual - gasto_mensual
ahorro_anual = ahorro_mensual * 12
```

```python
# 4. Generar reporte completo usando F-strings

reporte_personal = f"""

===============================================
REPORTE PERSONAL                     
===============================================
                                                      
INFORMACIÓN PERSONAL                             
    Nombre: {nombre}                                 
    Edad: {edad} años                                
    Peso: {peso} kg                                  
    Altura: {altura} m                               
    IMC: {imc:.1f}                                   

------------------------------------------------

INFORMACIÓN FINANCIERA                           
    Ingreso mensual: ${ingreso_mensual:,.2f}         
    Gasto mensual: ${gasto_mensual:,.2f}             
    Ahorro mensual: ${ahorro_mensual:,.2f}           
    Ahorro anual proyectado: ${ahorro_anual:,.2f}   

===============================================
"""

print(reporte_personal)
```
