```python
# En Python se emplean condicionales para controloar el flujo de ejecución
# de un programa o programas. La estructural de control más simple es 'if'

# Ejecutar un código solo si una condición es verdadera (True)
if True:
    print("Este código se ejecuta")

# No ejecutar un código si una condición es falsa (False)
if False:
    print("Este código nunca se ejecutará")
```

```python
# Verificar si una persona es mayor de edad para acceder a un sitio
edad_usuario = 20

# La palabra 'if' inicia la condición
# La condición es evaluar si la edad del usuario es mayor o igual a 18
# Es OBLIGATORIO colocar ':' al final de la línea de código dle 'if'

if edad_usuario >= 18:
    print("Acceso permitido. Bienvenido/a al sitio.")

print("El programa ha terminado.")
```

```python
# Podemos combinar input() con condicionales para ejecutar un código
# Supongamos queremos validar que una constraseña sea correcta

# La contraseña correcta se guarda en una variable
constraseña_corecta = "Python18!!"

# Pedimos al usuario que ingrese la contraseña
contraseña_ingresada = input("Por favor, introduce la contraseña:")

# Comparamos las contraseñas y validamos
if constraseña_corecta == contraseña_ingresada:
    print("¡Contraseña correcta! Has iniciado sesión.")

# Si la contraseña fuera incorrecta, el programa simplemente no ejecuta el print() y 
# finaliza la ejecución del programa. Para estos casos, se soluciona usando 
# un statement llamado 'else'
```

```python
# Podemos emplear 'if' con el statement 'else' para ejecutar un código
# de manera alternativa cuando nuestra condición no se cumpla
# Vamos a suponer que queremos evaluar si un estudiante aprobó un examen

# Calificación del examen
calif_minimina = 6.5
calif_alumno = float(input("Introduce la calificación del estudiante:"))

# Evaluamos con condicionales
if calif_alumno >= calif_minimina:
    print("¡Felicidades! Has aprobado.")
else:
    print("Lo sentimos, no has aprobado")

print("Revisión de calificaciones terminada.")
```

```python
# Emplear variables y operadores lógicos para controlar el flujo de ejecución
# Supongamos que tenemos un presupuesto para comprar artículos
presupuesto = 15000

# Artículos
art1 = 1200
art2 = 10000
art3 = 5000
art4 = 6500
art5 = 3700

# Flujo de ejecución al comprar articulos 1 y 2
if (art1 + art2) < presupuesto:
    print("Esta compra está dentro del presupuesto")

# Flujo de ejecución al comprar artículos 1, 2 y 3
if (art1 + art2 + art3) < presupuesto:
    print("Esta compra está dentro del presupuesto")
else:
    print("Esta compra está fuera del presupuesto")

# --- Añadir otro condicional ----
# Suma de los artículos
costo_total = art2 + art3

# Flujo de ejecución
if costo_total < presupuesto:
    print("Esta compra está dentro del presupuesto")
elif costo_total <= presupuesto:
    print("Esta compra es igual al presupuesto")
else:
    print("Esta compra está fuera del presupuesto")
```

```python
# Podemos evaluar más de una condición al mismo tiempo, esto se logra
# utilizando el operador 'and'

# Simulación de una base de datos de un usuario
usuario_registrado = "admin"
constraseña_usuario = "pass-admin-2025"

# Solicitamos datos del usuario
usuario_ingresado = input("Usuario: ")
contraseña_ingresada = input("Contraseña: ")

# La condición exige que ambas comparaciones sean True
if usuario_registrado == usuario_ingresado and constraseña_usuario == contraseña_ingresada:
    print("Acceso concedido. Bienvenido/a.")
else:
    print("Acceso denegado. Usuario o contraseña incorrectos.")
```

```python
# Podemos evaluar más de una condición y obtener un resultado si 
# al menos una de ellas se cumple con el operador 'or'

# Simulación de un descuento para estudiantes (sin importar la edad)
edad_usuario = int(input("Ingresa tu edad: "))
es_estudiante = bool(input("¿Eres estudiante activo?: (1 - Yes; 0 - No)"))

# La condición exige que solo se cumpla una
if edad_usuario < 18 or es_estudiante == True:
    print("Eres elegible para el descuento.")
else:
    print("No eres elegible para el descuento.")
```

```python
# Podemos combinar multiples if para generar if anidados
# Supongamos que queremos evaluar si un número es positivo, negativo o cero
numero = 5
if numero > 0:
    print("El número es positivo.")
else:
    if numero < 0:
        print("El número es negativo.")
    else:
        print("El número es cero.")

# Otra forma de escribir if anidados es usando 'elif'
# Supongamos que queremos evaluar si un número es positivo, negativo o cero
numero = 5
if numero > 0:
    print("El número es positivo.")
elif numero < 0:
    print("El número es negativo.")
else:
    print("El número es cero.")
```

```python
# Podemos usar list comprehension con condicionales para crear listas
# Supongamos que queremos crear una lista de números pares del 1 al 20
numeros_pares = [num for num in range(1, 21) if num % 2 == 0]
print(numeros_pares)
```
