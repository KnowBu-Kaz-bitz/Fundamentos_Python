# Fundamentos de Python: Proyecto Integrador

## Semana 2 - Sistema de Gestión de Calificaciones

### Objetivo del proyecto:
Crear un sistema completo que combine:
- Condicionales (if/elif/else)
- Bucles (for/while)
- Variables y operadores
- Entrada de usuario interactiva

### Descripción:
Desarrollaremos un sistema para gestionar calificaciones de estudiantes que permita ingresar datos, procesarlos y generar reportes automáticos.

---

```python
# PROYECTO INTEGRADOR: Sistema de Gestión de Calificaciones

print("="*50)
print("   SISTEMA DE GESTIÓN DE CALIFICACIONES")
print("="*50)

# Fase 1: Recopilación de datos básicos
nombre_curso = input("\n¿Cuál es el nombre del curso? ")
nombre_profesor = input("¿Cuál es el nombre del profesor/a? ")
num_estudiantes = int(input("¿Cuántos estudiantes hay en el curso? "))

print(f"\nConfiguración inicial completada:")
print(f"   Curso: {nombre_curso}")
print(f"   Profesor/a: {nombre_profesor}")
print(f"   Estudiantes: {num_estudiantes}")

# Listas para almacenar información
nombres_estudiantes = []
calificaciones = []

print(f"\nAhora ingresemos los datos de cada estudiante:")
```

```python
# Fase 2: Ingreso de datos con bucles y validación

# Usar bucle for para ingresar datos de cada estudiante
for i in range(num_estudiantes):
    print(f"\n--- Estudiante #{i+1} ---")
    
    nombre = input("Nombre del estudiante: ")
    
    # Validar que la calificación esté en rango válido
    while True:
        try:
            calificacion = float(input("Calificación (0-10): "))
            if 0 <= calificacion <= 10:
                break
            else:
                print(" La calificación debe estar entre 0 y 10")
        except ValueError:
            print(" Por favor ingresa un número válido")
    
    # Almacenar datos
    nombres_estudiantes.append(nombre)
    calificaciones.append(calificacion)
    
    print(f"{nombre}: {calificacion} - Registrado exitosamente")

print(f"\nDatos de {num_estudiantes} estudiantes registrados correctamente")
```

```python
# Fase 3: Procesamiento de datos y cálculos

print("\n" + "="*50)
print("           PROCESANDO CALIFICACIONES")
print("="*50)

# Cálculos estadísticos básicos
suma_total = 0
calificacion_mayor = calificaciones[0]
calificacion_menor = calificaciones[0]
estudiantes_aprobados = 0
estudiantes_reprobados = 0

# Procesar cada calificación
for i in range(len(calificaciones)):
    calificacion = calificaciones[i]
    suma_total += calificacion
    
    # Encontrar mayor y menor
    if calificacion > calificacion_mayor:
        calificacion_mayor = calificacion
        estudiante_mayor = nombres_estudiantes[i]
    
    if calificacion < calificacion_menor:
        calificacion_menor = calificacion
        estudiante_menor = nombres_estudiantes[i]
    
    # Contar aprobados y reprobados
    if calificacion >= 6.0:
        estudiantes_aprobados += 1
    else:
        estudiantes_reprobados += 1

# Calcular promedio
promedio_grupo = suma_total / len(calificaciones)

print("Cálculos completados exitosamente")
```

```python
# Fase 4: Generación de reporte detallado con clasificaciones

print("\n" + "="*60)
print("              REPORTE FINAL DE CALIFICACIONES")
print("="*60)

print(f"""
INFORMACIÓN DEL CURSO:
   Curso: {nombre_curso}
   Profesor/a: {nombre_profesor}
   Total de estudiantes: {num_estudiantes}

ESTADÍSTICAS GENERALES:
   Promedio del grupo: {promedio_grupo:.2f}
   Calificación más alta: {calificacion_mayor} ({estudiante_mayor})
   Calificación más baja: {calificacion_menor} ({estudiante_menor})
   Estudiantes aprobados: {estudiantes_aprobados}
   Estudiantes reprobados: {estudiantes_reprobados}
   Porcentaje de aprobación: {(estudiantes_aprobados/num_estudiantes)*100:.1f}%
""")

print("DETALLE POR ESTUDIANTE:")
print("-" * 60)

for i in range(len(nombres_estudiantes)):
    nombre = nombres_estudiantes[i]
    calificacion = calificaciones[i]
    
    # Clasificar el rendimiento
    if calificacion >= 9.0:
        categoria = "EXCELENTE"
    elif calificacion >= 8.0:
        categoria = "MUY BUENO"
    elif calificacion >= 7.0:
        categoria = "BUENO"
    elif calificacion >= 6.0:
        categoria = "SUFICIENTE"
    else:
        categoria = "INSUFICIENTE"

    # Determinar si está arriba o abajo del promedio
    if calificacion > promedio_grupo:
        posicion = "Arriba del promedio"
    elif calificacion < promedio_grupo:
        posicion = "Debajo del promedio" 
    else:
        posicion = "En el promedio"
    
    print(f"{i+1:2d}. {nombre:20s} | {calificacion:4.1f} | {categoria:15s} | {posicion}")

print("="*60)
print(f"Reporte generado para el curso '{nombre_curso}'")
print("="*60)
```
