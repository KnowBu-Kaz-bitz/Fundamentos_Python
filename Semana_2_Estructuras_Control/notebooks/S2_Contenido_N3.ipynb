## Semana 2 - Sistema de Gesti√≥n de Calificaciones

### Objetivo del proyecto:
Crear un sistema completo que combine:
- Condicionales (if/elif/else)
- Bucles (for/while)
- Variables y operadores
- Entrada de usuario interactiva

### Descripci√≥n:
Desarrollaremos un sistema para gestionar calificaciones de estudiantes que permita ingresar datos, procesarlos y generar reportes autom√°ticos.

---

```python
# PROYECTO INTEGRADOR: Sistema de Gesti√≥n de Calificaciones

print("="*50)
print("   SISTEMA DE GESTI√ìN DE CALIFICACIONES")
print("="*50)

# Fase 1: Recopilaci√≥n de datos b√°sicos
nombre_curso = input("\n¬øCu√°l es el nombre del curso? ")
nombre_profesor = input("¬øCu√°l es el nombre del profesor/a? ")
num_estudiantes = int(input("¬øCu√°ntos estudiantes hay en el curso? "))

print(f"\n‚úÖ Configuraci√≥n inicial completada:")
print(f"   Curso: {nombre_curso}")
print(f"   Profesor/a: {nombre_profesor}")
print(f"   Estudiantes: {num_estudiantes}")

# Listas para almacenar informaci√≥n
nombres_estudiantes = []
calificaciones = []

print(f"\nüìù Ahora ingresemos los datos de cada estudiante:")
```

```python
# Fase 2: Ingreso de datos con bucles y validaci√≥n

# Usar bucle for para ingresar datos de cada estudiante
for i in range(num_estudiantes):
    print(f"\n--- Estudiante #{i+1} ---")
    
    nombre = input("Nombre del estudiante: ")
    
    # Validar que la calificaci√≥n est√© en rango v√°lido
    while True:
        try:
            calificacion = float(input("Calificaci√≥n (0-10): "))
            if 0 <= calificacion <= 10:
                break
            else:
                print("‚ùå La calificaci√≥n debe estar entre 0 y 10")
        except ValueError:
            print("‚ùå Por favor ingresa un n√∫mero v√°lido")
    
    # Almacenar datos
    nombres_estudiantes.append(nombre)
    calificaciones.append(calificacion)
    
    print(f"‚úÖ {nombre}: {calificacion} - Registrado exitosamente")

print(f"\nüéØ Datos de {num_estudiantes} estudiantes registrados correctamente")
```

```python
# Fase 3: Procesamiento de datos y c√°lculos

print("\n" + "="*50)
print("           PROCESANDO CALIFICACIONES")
print("="*50)

# C√°lculos estad√≠sticos b√°sicos
suma_total = 0
calificacion_mayor = calificaciones[0]
calificacion_menor = calificaciones[0]
estudiantes_aprobados = 0
estudiantes_reprobados = 0

# Procesar cada calificaci√≥n
for i in range(len(calificaciones)):
    calificacion = calificaciones[i]
    suma_total += calificacion
    
    # Encontrar mayor y menor
    if calificacion > calificacion_mayor:
        calificacion_mayor = calificacion
        estudiante_mayor = nombres_estudiantes[i]
    
    if calificacion < calificacion_menor:
        calificacion_menor = calificacion
        estudiante_menor = nombres_estudiantes[i]
    
    # Contar aprobados y reprobados
    if calificacion >= 6.0:
        estudiantes_aprobados += 1
    else:
        estudiantes_reprobados += 1

# Calcular promedio
promedio_grupo = suma_total / len(calificaciones)

print("üìä C√°lculos completados exitosamente")
```

```python
# Fase 4: Generaci√≥n de reporte detallado con clasificaciones

print("\n" + "="*60)
print("              REPORTE FINAL DE CALIFICACIONES")
print("="*60)

print(f"""
üìö INFORMACI√ìN DEL CURSO:
   Curso: {nombre_curso}
   Profesor/a: {nombre_profesor}
   Total de estudiantes: {num_estudiantes}

üìä ESTAD√çSTICAS GENERALES:
   Promedio del grupo: {promedio_grupo:.2f}
   Calificaci√≥n m√°s alta: {calificacion_mayor} ({estudiante_mayor})
   Calificaci√≥n m√°s baja: {calificacion_menor} ({estudiante_menor})
   Estudiantes aprobados: {estudiantes_aprobados}
   Estudiantes reprobados: {estudiantes_reprobados}
   Porcentaje de aprobaci√≥n: {(estudiantes_aprobados/num_estudiantes)*100:.1f}%
""")

print("üë• DETALLE POR ESTUDIANTE:")
print("-" * 60)

for i in range(len(nombres_estudiantes)):
    nombre = nombres_estudiantes[i]
    calificacion = calificaciones[i]
    
    # Clasificar el rendimiento
    if calificacion >= 9.0:
        categoria = "EXCELENTE üåü"
    elif calificacion >= 8.0:
        categoria = "MUY BUENO üëç"
    elif calificacion >= 7.0:
        categoria = "BUENO ‚úì"
    elif calificacion >= 6.0:
        categoria = "SUFICIENTE ‚ö™"
    else:
        categoria = "INSUFICIENTE ‚ùå"
    
    # Determinar si est√° arriba o abajo del promedio
    if calificacion > promedio_grupo:
        posicion = "Arriba del promedio"
    elif calificacion < promedio_grupo:
        posicion = "Debajo del promedio" 
    else:
        posicion = "En el promedio"
    
    print(f"{i+1:2d}. {nombre:20s} | {calificacion:4.1f} | {categoria:15s} | {posicion}")

print("="*60)
print(f"Reporte generado para el curso '{nombre_curso}'")
print("="*60)
```
