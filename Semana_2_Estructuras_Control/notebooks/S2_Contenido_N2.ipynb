```python
# En Python existen los loops o bucles, que permiten ejecutar un bloque de código múltiples veces de 
# manera eficiente. Los dos tipos principales de bucles en Python son "for" y "while".

# Bucle "for"
# El bucle "for" se utiliza para iterar sobre una secuencia (como una lista, tupla o cadena de texto).
# Aquí hay un ejemplo de cómo usar un bucle "for" en Python:
for i in range(5):
    print("Iteración:", i)

# Bucle "while"
# El bucle "while" se utiliza para repetir un bloque de código mientras se cumpla una condición.
# Aquí hay un ejemplo de cómo usar un bucle "while" en Python:
contador = 0
while contador < 5:
    print("Contador:", contador)
    contador += 1
```

```python
# Podemos combinar loops con estructuras de control como "if" para crear lógica más compleja.
# Aquí hay un ejemplo que utiliza un bucle "for" junto con una estructura "if":
for numero in range(10):
    if numero % 2 == 0:
        print(numero, "es un número par")
    else:
        print(numero, "es un número impar")
```

```python
# Combinando bucles anidados
# También podemos anidar bucles para crear estructuras más complejas. Aquí hay un ejemplo:
for i in range(3):
    for j in range(2):
        print("i:", i, "j:", j)
```

```python
# Podemos crear bucles más complejos según nuestras necesidades, 
# combinando diferentes tipos de bucles y estructuras de control.
# Algunos ejemplos adicionales incluyen el uso de "break" y "continue" para controlar el flujo del bucle.
# "break" se utiliza para salir de un bucle antes de que termine, 
# mientras que "continue" se utiliza para saltar a la siguiente iteración del bucle.
# Aquí hay un ejemplo que utiliza "break" y "continue":
for i in range(10):
    if i == 5:
        break  # Salir del bucle cuando i es igual a 5
    if i % 2 == 0:
        continue  # Saltar los números pares
    print("Número impar menor que 5:", i)
```

```python
# Otros ejemplos más complejos son los siguientes:
# 1. Bucle que calcula la suma de los primeros n números naturales.
n = 10
suma = 0
for i in range(1, n + 1):
    suma += i
print("La suma de los primeros", n, "números naturales es:", suma)
```

```python
# 2. Bucle que encuentra el máximo en una lista de números.
numeros = [3, 5, 2, 8, 1]
maximo = numeros[0]
for num in numeros:
    if num > maximo:
        maximo = num
print("El número máximo en la lista es:", maximo)
```

```python
# 3. Bucle que genera la tabla de multiplicar de un número dado.
numero = 7
for i in range(1, 11):
    print(numero, "x", i, "=", numero * i)
```

```python
# Ejemplo práctico: evaluar si un año es bisiesto utilizando un bucle y una estructura condicional.
años = [1996, 2000, 2001, 2004, 1900, 2020]
for año in años:
    if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
        print(año, "es un año bisiesto.")
    else:
        print(año, "no es un año bisiesto.")
```
