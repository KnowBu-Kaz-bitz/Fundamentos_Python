### ¬øPor qu√© son importantes los operadores?
Los operadores son las herramientas b√°sicas para realizar c√°lculos, comparaciones y tomar decisiones en Python. Son la base de toda la l√≥gica de programaci√≥n.

```python
# Ejercicio 3.1: Operadores aritm√©ticos b√°sicos

print("=== OPERADORES ARITM√âTICOS EN PYTHON ===\n")

# Definir dos n√∫meros para trabajar
a = 15
b = 4

print(f"Trabajando con: a = {a}, b = {b}\n")

# Operadores b√°sicos
print("üî¢ OPERACIONES B√ÅSICAS:")
print(f"   Suma: {a} + {b} = {a + b}")
print(f"   Resta: {a} - {b} = {a - b}")
print(f"   Multiplicaci√≥n: {a} * {b} = {a * b}")
print(f"   Divisi√≥n: {a} / {b} = {a / b:.2f}")

print(f"\nüîç OPERACIONES ESPECIALES:")
print(f"   Divisi√≥n entera: {a} // {b} = {a // b}")
print(f"   M√≥dulo (residuo): {a} % {b} = {a % b}")
print(f"   Exponenciaci√≥n: {a} ** {b} = {a ** b}")

# Ejemplo pr√°ctico: calcular el √°rea de un c√≠rculo
import math
radio = 5
area = math.pi * radio ** 2
print(f"\nüåç EJEMPLO PR√ÅCTICO:")
print(f"   √Årea de c√≠rculo con radio {radio}: {area:.2f}")
```

```python
# Ejercicio 3.2: Precedencia de operadores y operadores de asignaci√≥n

print("=== PRECEDENCIA DE OPERADORES ===\n")

# Demostrar precedencia (orden de operaciones)
resultado1 = 2 + 3 * 4  # Sin par√©ntesis
resultado2 = (2 + 3) * 4  # Con par√©ntesis

print("üìê ORDEN DE OPERACIONES:")
print(f"   2 + 3 * 4 = {resultado1} (primero multiplicaci√≥n, luego suma)")
print(f"   (2 + 3) * 4 = {resultado2} (primero par√©ntesis, luego multiplicaci√≥n)")

# Ejemplo m√°s complejo
complejo = 10 + 2 ** 3 * 4 - 5
print(f"   10 + 2 ** 3 * 4 - 5 = {complejo}")
print("   Orden: 2**3=8, 8*4=32, 10+32=42, 42-5=37")

print(f"\nüîÑ OPERADORES DE ASIGNACI√ìN COMBINADOS:")
x = 10
print(f"   x inicial: {x}")

x += 5  # Equivale a: x = x + 5
print(f"   x += 5 ‚Üí x = {x}")

x *= 2  # Equivale a: x = x * 2  
print(f"   x *= 2 ‚Üí x = {x}")

x //= 3  # Equivale a: x = x // 3
print(f"   x //= 3 ‚Üí x = {x}")

x **= 2  # Equivale a: x = x ** 2
print(f"   x **= 2 ‚Üí x = {x}")

print("\nüí° Estos operadores hacen el c√≥digo m√°s conciso y legible")
```

## 2. Familiarizaci√≥n con operadores l√≥gicos
#### Objetivos del notebook.
- Objetivo 1: Conocer los diferentes operadores l√≥gicos en Python
- Objetivo 2: Identificar la prioridad de operadores y sus usos
- Dataset: No se requieren ning√∫n archivo adicional

```python
# Ejercicio 3.3: Operadores l√≥gicos y de comparaci√≥n completos

print("=== OPERADORES L√ìGICOS Y DE COMPARACI√ìN ===\n")

# Variables para ejemplos
edad = 25
es_estudiante = True
tiene_trabajo = True
salario = 30000

print(f"Datos de ejemplo:")
print(f"   Edad: {edad}")
print(f"   Es estudiante: {es_estudiante}")
print(f"   Tiene trabajo: {tiene_trabajo}")
print(f"   Salario: ${salario:,}")

print(f"\nüîó OPERADORES DE COMPARACI√ìN:")
print(f"   edad == 25: {edad == 25}")
print(f"   edad != 30: {edad != 30}")
print(f"   edad >= 18: {edad >= 18}")
print(f"   edad < 65: {edad < 65}")
print(f"   salario > 25000: {salario > 25000}")

print(f"\nüß† OPERADORES L√ìGICOS:")
print(f"   es_estudiante AND tiene_trabajo: {es_estudiante and tiene_trabajo}")
print(f"   edad >= 18 OR es_estudiante: {edad >= 18 or es_estudiante}")
print(f"   NOT es_estudiante: {not es_estudiante}")

# Ejemplo pr√°ctico: elegibilidad para pr√©stamo
print(f"\nüí≥ EJEMPLO: ELEGIBILIDAD PARA PR√âSTAMO")
edad_minima = edad >= 18
ingresos_suficientes = salario >= 25000
tiene_empleo = tiene_trabajo

elegible = edad_minima and ingresos_suficientes and tiene_empleo
print(f"   ¬øElegible para pr√©stamo? {elegible}")
print(f"   Raz√≥n: Edad OK({edad_minima}) + Ingresos OK({ingresos_suficientes}) + Empleo({tiene_empleo})")

# Operadores de pertenencia
frutas = ['manzana', 'banana', 'naranja']
print(f"\nüîç OPERADORES DE PERTENENCIA:")
print(f"   'banana' in {frutas}: {'banana' in frutas}")
print(f"   'uva' not in {frutas}: {'uva' not in frutas}")
```

```python
# Ejercicio 3.4: Proyecto integrador - Calculadora de finanzas personales

print("="*60)
print("           CALCULADORA DE FINANZAS PERSONALES")
print("="*60)

# Datos de entrada
print("\nüí∞ INGRESOS MENSUALES:")
salario_base = 25000
bonos = 3000
ingresos_extra = 1500
ingresos_totales = salario_base + bonos + ingresos_extra

print(f"   Salario base: ${salario_base:,}")
print(f"   Bonos: ${bonos:,}")
print(f"   Ingresos extra: ${ingresos_extra:,}")
print(f"   Total ingresos: ${ingresos_totales:,}")

# Gastos mensuales
print(f"\nüí∏ GASTOS MENSUALES:")
renta = 8000
comida = 4000
transporte = 2000
entretenimiento = 1500
servicios = 1200
gastos_totales = renta + comida + transporte + entretenimiento + servicios

print(f"   Renta: ${renta:,}")
print(f"   Comida: ${comida:,}")
print(f"   Transporte: ${transporte:,}")
print(f"   Entretenimiento: ${entretenimiento:,}")
print(f"   Servicios: ${servicios:,}")
print(f"   Total gastos: ${gastos_totales:,}")

# An√°lisis financiero usando operadores
dinero_libre = ingresos_totales - gastos_totales
porcentaje_gastos = (gastos_totales / ingresos_totales) * 100
porcentaje_ahorro = (dinero_libre / ingresos_totales) * 100

# Proyecciones anuales
ahorro_anual = dinero_libre * 12
gastos_anuales = gastos_totales * 12

print(f"\nüìä AN√ÅLISIS FINANCIERO:")
print(f"   Dinero libre mensual: ${dinero_libre:,}")
print(f"   Porcentaje de gastos: {porcentaje_gastos:.1f}%")
print(f"   Porcentaje de ahorro: {porcentaje_ahorro:.1f}%")
print(f"   Ahorro proyectado anual: ${ahorro_anual:,}")

# Evaluaciones con operadores l√≥gicos
print(f"\nüéØ EVALUACI√ìN DE SALUD FINANCIERA:")
gastos_controlados = porcentaje_gastos <= 70
ahorro_saludable = porcentaje_ahorro >= 20
situacion_estable = dinero_libre > 0

print(f"   ‚úì Gastos controlados (<70%): {gastos_controlados}")
print(f"   ‚úì Ahorro saludable (>=20%): {ahorro_saludable}")
print(f"   ‚úì Situaci√≥n estable (>$0): {situacion_estable}")

salud_financiera = gastos_controlados and ahorro_saludable and situacion_estable

if salud_financiera:
    print(f"\nüåü ¬°EXCELENTE! Tu salud financiera es muy buena.")
elif situacion_estable and gastos_controlados:
    print(f"\nüëç BUENO: Situaci√≥n estable, pero puedes ahorrar m√°s.")
elif situacion_estable:
    print(f"\n‚ö†Ô∏è REGULAR: Controla mejor tus gastos.")
else:
    print(f"\n‚ùå CR√çTICO: Gastas m√°s de lo que ganas.")

print("="*60)
```
