## 1. Familiarización con la sintaxis de funciones simples
#### Objetivos del notebook.
- Objetivo 1: Conocer la estructura de una función
- Objetivo 2: Identificar los parámetros de una función
- Dataset: No se requieren ningún archivo adicional

```python
# Una función se define utilizando la palabra clave `def`, seguida del nombre de la función y paréntesis
def saludar(nombre):
    # El cuerpo de la función está indentado
    mensaje = f"Hola, {nombre}!"
    return mensaje  # La función devuelve un valor con la palabra clave `return`
```

```python
# Llamada a la función
resultado = saludar("César")
print(resultado)  # Imprime el resultado de la función
```

```python
# Una función puede tener múltiples parámetros separados por comas
def sumar(a, b):
    return a + b
```

```python
# Llamada a la función
resultado = sumar(5, 3) 
print(resultado)  # Imprime el resultado de la función
```

```python
# Usamos return y no print dentro de las funciones para devolver valores
# Al usar print dentro de una función, simplemente mostramos el valor en la consola, pero no lo devolvemos para su uso posterior.
```

## 2. Uso de control flow para definir condiciones en funciones
#### Objetivos del notebook.
- Objetivo 1: Complementar el uso de funciones con control flow
- Objetivo 2: Combinar lo aprendido previamente
- Dataset: No se requieren ningún archivo adicional

```python
# Podemos combinar control flow dentro de funciones
def evaluar_numero(num):
    if num > 0:
        return "Positivo"
    elif num < 0:
        return "Negativo"
    else:
        return "Cero"
```

```python
# llamada a la función
print(evaluar_numero(10))  # Imprime "Positivo"
```

```python
# Podemos crear funciones más complejas combinando múltiples conceptos aprendidos previamente.
def procesar_numero(num):
    evaluacion = evaluar_numero(num)
    if evaluacion == "Positivo":
        return f"El número {num} es positivo."
    elif evaluacion == "Negativo":
        return f"El número {num} es negativo."
    else:
        return f"El número {num} es cero."
    
# llamada a la función
print(procesar_numero(-5))  # Imprime "El número -5 es negativo."
```

```python
# Podemos definir funciones con una lógica más compleja usando operadores lógicos y aritméticos,
# estructuras de control y bucles.
def es_bisiesto(año):
    return (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0)
# llamada a la función
print(es_bisiesto(2020))  # Imprime True
```
